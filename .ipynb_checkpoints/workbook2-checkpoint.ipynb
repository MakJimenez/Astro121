```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting data 

importdata = np.load("testfreq1420.npz")
indexdata = importdata["arr_1"]
print(importdata["arr_0"])
data = indexdata[3]

# Convert I/Q to complex data
complexdata = data[:,0] + 1j * data[:,1]

# FFT Parameters 
sampling_rate = 2.9e6
N = len(complexdata)

# Hann window + FFT
window = np.hanning(N)
fft_data = np.fft.fftshift(np.fft.fft(complexdata * window))

# Power Spectrum
power = np.abs(fft_data)**2
power_db = 10 * np.log10(power)

# Frequency axis (baseband)
f = np.fft.fftshift(np.fft.fftfreq(N, d=1/sampling_rate))

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(f / 1e3, power_db)
plt.axvline(-155, linestyle=":", color="black", linewidth=2) # expected signal location
plt.yscale("log")
plt.xlim(-156, -154)
plt.xlabel("Frequency Offset (kHz)")
plt.ylabel("Power (dB)")
plt.title("Basebamd Power Spectrum (I/Q Data, Hann Window)")
plt.grid(True)
plt.show()
```

```python
# Averaging over all the power spectra (4) to try and reduce noise 

# amplitude to zero dBm, frequency 1420.405 MHz

# Importing data 
importdata = np.load("freq1420.405.npz")
indexdata = importdata["arr_1"]
print(importdata["arr_0"])
data = indexdata[3]

# Convert I/Q to complex data
complexdata = data[:,0] + 1j * data[:,1]

# FFT Parameters 
sampling_rate = 2.9e6
N = len(complexdata)
navg = 4

# Hann window + FFT
window = np.hanning(N)
fft_data = np.fft.fftshift(np.fft.fft(complexdata * window))

# Power Spectrum
power = np.abs(fft_data)**2
power_db = 10 * np.log10(power)

# Frequency axis (baseband)
f = np.fft.fftshift(np.fft.fftfreq(N, d=1/sampling_rate))

# Loop over captures
power_accum = np.zeros(N)

for i in range(1, navg):
        d = data[i]
#         x = d[:,0] + 1j * d[:,1] # I/Q -> complex
        
        fft_data = np.fft.fftshift(np.fft.fftshift(complexdata * window))
        power_accum += np.abs(fft_data)**2

# Average + convert to dB
power_avg = power_accum / navg
power_avg_db = 10 * np.log10(power_avg)
        
# Plotting
plt.figure(figsize=(10, 4))
plt.plot(f / 1e3, power_avg_db, color="teal")
plt.axvline(10, linestyle=":", color="black", linewidth=2) # expected signal location
# plt.yscale("log")
# plt.xlim(-25, 25)
# plt.xlim()
plt.xlabel("Frequency Offset (kHz)")
plt.ylabel("Power (dB)")
plt.title(f"Baseband Power Spectrum ({navg} averages)")
plt.grid(True)
plt.show()
```
